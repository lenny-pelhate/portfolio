### **Change of probability**

**Radon-Nikodym derivative process**
<br>
- $D$ est une $(\mathcal{F_t}, \mathbb{P})$-martingale u.i.
- $D$ admet une modification càdlàg, pour cette version et pour tout temps d'arrêt $T$, on a $D_T = \frac{d\mathbb{Q}}{d\mathbb{P}}\big|_{\mathcal{F}_T}$
- Si $\mathbb{Q} \sim \mathbb{P}$ sur $\mathcal{F}_\infty$ (i.e., $\mathbb{Q} \ll \mathbb{P}$ et $\mathbb{P} \ll \mathbb{Q}$), alors p.s. $\forall t \ge 0, D_t > 0$

Sur $[0,T]$, $D_t = \frac{d\mathbb{Q}}{d\mathbb{P}}\big|_{\mathcal{F}_t} = \mathbb{E}_{\mathbb{P}} \Big[D_T | \mathcal{F}_t \Big]$ p.s. par unicité de la dérivée de Radon-Nikodym. $D$ est une $\mathcal{F}_t$ martingale u.i. car fermée.

Logarithme Stochastique:
<br>
Soit D une martingale locale à trajectoires continues strictement positive. Alors, il existe une unique martingale locale, à trajectoires continues, $L$, appelée logarithme stochastique de $D$, tq:
$$
D_t = exp \Big(L_t - \frac{1}{2} \langle L, L \rangle _t \Big) = \mathcal{E}(L)_t
$$ 

Soit $\mathbb{Q} \sim \mathbb{P}$ sur $\mathcal{F}_\infty$  et $L$ le logarithme stochastique associé à la martingale $D_t = \mathbb{E}_{\mathbb{P}} [D_{\infty} | \mathcal{F}_t]$ qu'on suppose à trajectoires continues.
- Si $XD$ est une $\mathbb{P}$-martingale locale, alors $X$ est une $\mathbb{Q}$-martingale locale.

### **Change of Numeraire:**

A process $(N_t)_{t \ge 0}$ is a numeraire if:
- The discounted value of $N_t$ is a $\mathbb{P}$-martingale
- $(N_t)_{t \ge 0} > 0$
For all asset value $(X_t)_{t \ge 0}$ (the discounted value of $(X_t)_{t \ge 0}$ is a $\mathbb{P}$-martingale), $\Big(\frac{X_t}{N_t}\Big)$ is a martingale under $\mathbb{Q}$

A risk-neutral measure is a probability measure $\mathbb{P}$ under which the discounted asset price $(Y_t)_{t \ge 0} = (e^{-rt}X_t)_{t \ge 0}$ is a martingale
<br>
Thus, $\Big(e^{-\int_{0}^{t} r_s ds} X_t\Big)$ is a $\mathbb{P}$-martingale

>**Example 1:** (Savings account as the numeraire)
>
>Under $\mathbb{P}$:
>$$
>dS_t = \mu S_t dt + \sigma S_t dB^\mathbb{P}_t
>$$
>
>Let $N$ be the value of the savings account:
>$$
>N_T = e^{\int_0^T r_t \, dt} \qquad N_0 = 1
>$$
>
> Since $\Big(\frac{S_t}{N_t}\Big)_{t \ge 0}$ is a $\mathbb{Q}$-martingale:
>$$
> \frac{S_0}{N_0} = \mathbb{E}_{\mathbb{Q}} \bigg(\frac{S_T}{N_T} \bigg)
>$$
>
> which gives:
>$$
>S_0 = \mathbb{E}_{\mathbb{Q}} \bigg(S_T \, e^{-\int_0^T r_t \, dt} \bigg)
>$$
>
> With $\bar{r} = \frac{1}{T} \int_0^T r_t \, dt$, we have:
>$$
>S_0 = \mathbb{E}_{\mathbb{Q}} \bigg(S_T \, e^{-\bar{r} \, T} \bigg)
>$$
>
> If the interest rate $r$ is constant:
>$$
>S_0 = e^{-r \, T} \, \mathbb{E}_{\mathbb{Q}} (S_T)
>$$
>
><br>

>**Example 2:** (Zero-coupon as the numeraire)
>
>Under $\mathbb{P}$:
>$$
>dS_t = \mu S_t dt + \sigma S_t dB^\mathbb{P}_t
>$$
>
>Let B be the value of the ZCB:
>$$
>B(t,T) = \mathbb{E}_{\mathbb{P}} \bigg[e^{-\int_t^T r_s \, ds} \Big| \mathcal{F}_t \bigg] \qquad B(T,T) = 1
>$$
>
>Since $\Big(\frac{S_t}{B(t,T)}\Big)_{t \ge 0}$ is a martingale under the new probability $\mathbb{Q}^T$:
>$$
>\frac{S_0}{B(0,T)} = \mathbb{E}_{\mathbb{Q}^T} \bigg(\frac{S_T}{B(T,T)} \bigg)
>$$
>
>which gives:
>$$
>S_0 = B(0,T) \, \mathbb{E}_{\mathbb{Q}^T} (S_T)
>$$
>The key difference compared to the savings account is that here the discounting factor lies outside of the expected value.
>
>Note that $F(t,T) = \frac{S_t}{B(t,T)}$, where $F(t,T)$ is the forward price/rate observed at time $t$ for a contract maturing at time $T$. Thus the forward is a $\mathbb{Q}^T$-martingale, stating that the price of a forward contract maturing at time $T$ is the expected value of the future spot price at time $T$:
>$$
>F(0,T) = \mathbb{E}_{\mathbb{Q}^T} (S_T)
>$$
>
>For this reason, the probability measure $\mathbb{Q}^T$ is also called the **_forward-neutral measure_**.
>
> Also note that the forward price is different from the futures price. The futures price is the expected price at time T in the risk-neutral universe:
>$$
>F_{fut}(t,T) = \mathbb{E}_{\mathbb{Q}} (S_T)
>$$
> where $\mathbb{Q}$ is the risk-neutral measure.
>
><br>

>**Example 3:** (Annuity as the numeraire)
>
>Under $\mathbb{P}$, we denote $V_t$ the value observed at time $t$, of the floating leg of a swap starting at time $T=T_0$, with payment dates at $\llbracket T_1 \, , T_N \rrbracket$.
>
>Let $A$ be the annuity of this swap:
>$$
>A_t = \sum_{i=0}^{n-1} (T_{i+1} - T_i) \, B(t,T_{i+1})
>$$
>
>Since $\Big(\frac{V_t}{A_t}\Big)_{t \ge 0}$ is a martingale under the new probability $\mathbb{Q}^S$:
>$$
>\frac{V_0}{A_0} = \mathbb{E}_{\mathbb{Q}^S} \bigg(\frac{V_T}{A_T} \bigg)
>$$
>
>which gives:
>$$
>V_0 = A_0 \, \mathbb{E}_{\mathbb{Q}^S} \bigg(\frac{V_T}{A_T} \bigg)
>$$
>
> Note that $s_t = \frac{V_t}{A_t}$, where $s_t$ if the forward swap rate, observed at time $t<T$. We can see that the swap rate $s_t$ is a $\mathbb{Q}^S$- martingale:
>$$
>s_0 = \mathbb{E}_{\mathbb{Q}^S} (s_T)
>$$
>
>For this reason, the probability measure $\mathbb{Q}^S$ is also called the **_swap-neutral measure_**.
>
><br>

### **Girsanov:**

**1-dimensional**
<br>
Let $\mathbb{Q} \sim \mathbb{P}$ on $\mathcal{F}_\infty$, and $L$ be the stochastic logarithm associated to the martingale $D_t = \frac{d\mathbb{Q}}{d\mathbb{P}}_{|\mathcal{F_t}}$.
<br>
If $M$ is a $(\mathcal{F_t}, \mathbb{P})$-local martingale with continuous paths, then $\tilde{M} = M - \langle M,L \rangle$ is a $(\mathcal{F_t}, \mathbb{Q})$-local martingale whose paths are continuous

cf Novikov's conditions

**n-dimensional**
<br>
Let $B$ be a $(\mathbb{P},\mathcal{F})$-brownian motion and let $\lambda$ be a n-dimensional $\mathcal{F}$-adapted process such that $\int_{0}^{t} ||\lambda_s|| < \infty$, a.s.
<br>
Define the local martingale $L$ as the solution of $dL_t = -L_t \lambda_t dB_t = -L_t \Big(\sum_i \lambda_t^i dB_t^i \Big)$, and $L_0=1$, so that:
$$
L_t = \exp \bigg(- \int_0^t \lambda_s dB_s - \frac{1}{2} \int_0^t ||\lambda_s||^2 ds \bigg)
$$
If $L$ is a martingale, the n-dimensional process
$$
\tilde{B}_t = B_t + \int_0^t \lambda_s ds \qquad \forall t \ge 0
$$
is a $\mathbb{Q}$-brownian motion, where $\mathbb{Q}|_{\mathcal{F}_t} = L_t \, \mathbb{P}|_{\mathcal{F}_t}$.

If $B$ is a brownian motion with correlation matrix $\Gamma$, then:
$$
\tilde{B}_t = B_t + \int_0^t \lambda_s \Gamma ds \qquad \forall t \ge 0
$$
is a $\mathbb{Q}$-brownian motion, with the same correlation matrix $\Gamma$


> **Example 1:** (from real-world probability $\mathbb{P}$ to risk-neutral probability $\mathbb{Q}$)
>
> Under $\mathbb{P}$:
> $$
> dS_t = \mu S_t dt + \sigma S_t dB^\mathbb{P}_t
> $$
>
> Let $L_t = \frac{r-\mu}{\sigma} B^\mathbb{P}_t$
> $$
> \langle B^\mathbb{P},L\rangle _t = \Big\langle B^\mathbb{P},\frac{r-\mu}{\sigma} B^\mathbb{P} \Big\rangle _t = \frac{r-\mu}{\sigma} \Big\langle B^\mathbb{P}, B^\mathbb{P} \Big\rangle _t = \frac{r-\mu}{\sigma} t
> $$
> 
> Thus, by Girsanov:
> $$
> B^\mathbb{Q}_t = B^\mathbb{P}_t - \langle B^\mathbb{P},L \rangle _t = \frac{\mu - r}{\sigma} t + B^\mathbb{P}_t
> $$
> 
> is a $(\mathcal{F_t}, \mathbb{Q})$ martingale and:
> \begin{align}
> dS_t &= \mu S_t dt + \sigma S_t dB^\mathbb{P}_t \\
>      &= \mu S_t dt + \sigma S_t \bigg(dB^\mathbb{Q}_t - \frac{\mu - r}{\sigma} dt\bigg) \\
>      &= r S_t dt + \sigma S_t dB^\mathbb{Q}_t \\
> \end{align}
> 
> which is a geometric brownian motion of drift r under the measure $\mathbb{Q}$
> 
> <br>

> **Example 2:** (from real-world probability $\mathbb{P}$ to forward-neutral probability $\mathbb{Q}^T$)
> 
> Under the real-world probability measure $\mathbb{P}$:
> $$
> dS_t = \mu_S \, S_t \, dt + \sigma_S \, S_t \, dW^\mathbb{P}_t
> $$ 
>
> Using a ZCB as the numeraire:
> 
> $$
> F(t,T) = \frac{S_t}{B(t,T)} 
> $$
>
> Under $\mathbb{P}$, the bond price process follows:
> $$
> dB(t,T) = r_t \, B(t,T) \, dt + \sigma_B(t,T) \, B(t,T) \, dW^\mathbb{P}_t
> $$
> 
> where $\sigma_B (t,T)$ is the lognormal volatility of $B(t,T)$
>
> Let $L_t = \sigma_B (t,T) \, W^\mathbb{P}_t$.
> 
> By Girsanov:
> $$
> W^{\mathbb{Q}^T}_t = W^\mathbb{P}_t - \langle W^\mathbb{P},L \rangle _t = W^\mathbb{P}_t - \sigma_B (t,T) \, t 
> $$
> 
> Applying Itô to $F(S,B) = \frac{S}{B}$:
> \begin{align}
> dF &= \frac{\partial F}{\partial S} dS + \frac{\partial F}{\partial B} dB + \frac{1}{2} \left[ \frac{\partial^2 F}{\partial S^2} dS^2 + 2 \frac{\partial^2 F}{\partial S \partial B} dS dB + \frac{\partial^2 F}{\partial B^2} dB^2 \right] \\
>    &= \frac{1}{B} dS - \frac{S}{B^2} dB - \frac{1}{B^2} dS dB + \frac{S}{B^3} dB^2\\
> \end{align}
> 
> Thus:
> \begin{align}
>\frac{dF}{F} &= \Big(\mu_S - r_t - \sigma_S \sigma_B + \sigma_B^2 \Big) dt + \Big(\sigma_S - \sigma_B \Big) dW_t^\mathbb{P} \\
> &= \Big(\mu_S - r_t - \sigma_S \sigma_B + \sigma_B^2 \Big) dt + \Big(\sigma_S - \sigma_B \Big) \Big(dW^{\mathbb{Q}^T}_t + \sigma_B \, dt \Big) \\
> &= \Big(\mu_S - r_t \Big) dt + \Big(\sigma_S - \sigma_B \Big) \, dW^{\mathbb{Q}^T}_t \\
> \end{align}
>
>Since in the forward measure $\mathbb{Q}^T$, the drift of tradable assets (like forward prices) must be zero, we set:
>$$
>\mu_S - r_t = 0
>$$
>
>Thus, under $\mathbb{Q}^T$, the dynamics simplify to:
>$$
>\frac{dF}{F} = (\sigma_S - \sigma_B) \, dW^{\mathbb{Q}^T}_t
>$$
>
>This shows that the forward price $F(t,T)$ follows a martingale under the forward-neutral measure $\mathbb{Q}^T$, and its lognormal volatility is $(\sigma_S - \sigma_B)$.
>
><br>


> **Example 3:** (from real-world probability $\mathbb{P}$ to swap-neutral probability $\mathbb{Q}^S$)
> 
> Under the real-world probability measure $\mathbb{P}$, the bond price process evolves according to:
> $$
> dB(t,T) = r_t \, B(t,T) \, dt + \sigma_B(t,T) \, B(t,T) \, dW^\mathbb{P}_t
> $$
>
> and the short rate process $r_t$ follows:
> $$
> dr_t = \mu_r(t,r_t) dt + \sigma_r(t,r_t) dW^\mathbb{P}_t
> $$
> 
> We want to apply Itô's lemma to the swap rate $s_t = \frac{V_t}{A_t} = \frac{B(t, T_0) - B(t, T_n)}{\sum_{j=1}^{n} (T_j - T_{j-1}) \, B(t, T_j)}$, but first:
> \begin{align}
> dV_t &= dB(t, T_0) - dB(t, T_n) \\
> &= \bigg(r_t \, B(t,T_0) \, dt + \sigma_B(t,T_0) \, B(t,T_0) \, dW^\mathbb{P}_t \bigg) - \bigg(r_t \, B(t,T_n) \, dt + \sigma_B(t,T_n) \, B(t,T_n) \, dW^\mathbb{P}_t \bigg) \\
> &= \Big(B(t,T_0) - B(t,T_n) \Big) r_t \, dt + \Big( \sigma_B(t,T_0) \, B(t,T_0) - \sigma_B(t,T_n) \, B(t,T_n) \Big) dW^\mathbb{P}_t \\
> &= V_t \bigg(r_t \, dt + \frac{\sigma_B(t,T_0) \, B(t,T_0) - \sigma_B(t,T_n) \, B(t,T_n)}{V_t} dW^\mathbb{P}_t \bigg)\\
> &= V_t \bigg(r_t \, dt + \sigma_V \, dW^\mathbb{P}_t \bigg) \\
> \end{align} 
>
> where $\sigma_V$ is the lognormal volatility of the floating leg of the swap, and:
> \begin{align}
> dA_t &= \sum_{i=1}^{n} (T_i - T_{i-1}) \, dB(t, T_i) \\
>       &= \sum_{i=1}^{n} (T_i - T_{i-1}) \, \bigg(r_t \, B(t,T) \, dt + \sigma_B(t,T) \, B(t,T) \, dW^\mathbb{P}_t \bigg) \\
>       &= A_t \bigg( r_t \, dt + \sum_{i=1}^{n} \frac{(T_i - T_{i-1}) \, \sigma_B(t, T_i) \, B(t, T_i)}{A_t} dW_t^{\mathbb{P}} \bigg) \\
>       &= A_t \bigg( r_t \, dt + \sigma_{s,t} \, dW_t^{\mathbb{P}} \bigg)
> \end{align}
> where $\sigma_{s,t}$ is the lognormal volatility of the annuity of the swap.
>
> Itô on applied to the swap rate $s = \frac{V}{A}$ gives:
> \begin{align}
> ds &= \frac{\partial s}{\partial V} dV + \frac{\partial s}{\partial A} dA + \frac{1}{2} \left[ \frac{\partial^2 s}{\partial V^2} dV^2 + 2 \frac{\partial^2 s}{\partial V \partial A} dV dA + \frac{\partial^2 s}{\partial A^2} dA^2 \right] \\
>    &= \frac{1}{A} dV - \frac{V}{A^2} dA - \frac{1}{A^2} dV dA + \frac{V}{A^3} dA^2 \\
> \end{align}
>
> Thus:
> \begin{align}
> \frac{ds}{s} &= (r_t \, dt + \sigma_V \, dW^\mathbb{P}_t) - (r_t \, dt + \sigma_{s,t} \, dW_t^{\mathbb{P}}) - \sigma_{s,t} \, \sigma_V \, dt + \sigma_{s,t}^2 \, dt\\
>  &= (\sigma_{s,t}^2 - \sigma_{s,t} \, \sigma_V) \, dt + (\sigma_V  - \sigma_{s,t}) \, dW_t^{\mathbb{P}} \\
> \end{align}
>
> Let $L_t = \sigma_{s,t} \, W^\mathbb{P}_t$.
>
> By Girsanov:
> $$
> W^{\mathbb{Q}^S}_t = W^\mathbb{P}_t - \langle W^\mathbb{P},L \rangle _t = W^\mathbb{P}_t - \sigma_{s,t} \, t 
> $$
>
> so:
> \begin{align}
> \frac{ds}{s} &= \Big(\sigma_{s,t}^2 - \sigma_{s,t} \, \sigma_V \Big) \, dt + \Big(\sigma_V  - \sigma_{s,t} \Big) \, \Big(dW^{\mathbb{Q}^S}_t + \sigma_{s,t} \, dt \Big) \\
>   &= \Big(\sigma_V  - \sigma_{s,t} \Big) \, dW^{\mathbb{Q}^S}_t
> \end{align}
>
>This shows that the swap rate $s$ follows a martingale under the swap-neutral measure $\mathbb{Q}^S$, and its lognormal volatility is $(\sigma_V - \sigma_{s,t})$.
>
